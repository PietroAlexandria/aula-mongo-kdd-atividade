4. Converter os resultados para DataFrame

```python
df = pd.DataFrame(list(db.atendimentos.find({}, {"_id": 0})))
print(f'DataFrame com {len(df)} registros e {len(df.columns)} colunas')
df.head()
```

5. Verificar valores nulos.

```python
print(df.isnull().sum())
```

6. Calcular faturamento total.

```python
fat_total = df['valor'].sum()
print(f'Faturamento total: R${fat_total:.2f}')
```

7. Calcular faturamento por cidade.

```python
fat_cidade = df.groupby('cidade')['valor'].sum().to_dict()
print(fat_cidade)
```

8. Calcular faturamento por serviço.

```python
fat_servico = df.groupby('servico')['valor'].sum().to_dict()
print(fat_servico)
```

9. Calcular ticket médio geral.

```python
ticket_medio_geral = df['valor'].mean()
print(f'Ticket médio geral: R${ticket_medio_geral:.2f}')
```

10. Calcular ticket médio por cidade.

```python
ticket_medio_por_cidade = df.groupby('cidade')['valor'].mean().to_dict()
print(ticket_medio_por_cidade)
```

11. Identificar o serviço mais lucrativo.

```python
servico_mais_lucrativo = max(fat_servico, key=fat_servico.get)
print(f'Serviço mais lucrativo: {servico_mais_lucrativo} (R${fat_servico[servico_mais_lucrativo]:.2f})')
```

12. Identificar a cidade com maior faturamento.

```python
cidade_maior_faturamento = max(fat_cidade, key=fat_cidade.get)
print(f'Cidade com maior faturamento: {cidade_maior_faturamento} (R${fat_cidade[cidade_maior_faturamento]:.2f})')
```

13. Identificar os 3 maiores atendimentos

```python
top3 = df.nlargest(3, 'valor')[['tutor', 'nome_pet', 'servico', 'valor', 'cidade']]
print(top3)
```
